```python
import numpy as np
from numpy import linalg, newaxis, random
#from matplotlib import collections, pyplot
import matplotlib.pyplot as plt
```

```python
def gen_rand_vecs(number, dims, p=2):
    vecs = random.normal(size=(number,dims))
    mags = linalg.norm(vecs, p, axis=-1)
    return vecs / mags[..., newaxis]
```

```python
fig, ax = plt.subplots(1,1, figsize=(10, 6))
#ax.set(xlim=(-1, 1), ylim=(-1, 1))
ax.axis("equal")
e = gen_rand_vecs(1000, 2, p=np.inf)
r = 1
e = e * r
start, end = min(e[:,0]), max(e[:,0])  + 0.1
ax.set(xlim=(start, end), ylim=(start, end))
ax.plot(e[:,0][..., newaxis], e[:,1][..., newaxis], ".r")
ax.text(0, 0, "center")
plt.show()
```
